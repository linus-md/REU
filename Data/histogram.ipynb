```julia
using CSV
using Plots
using DataFrames
```

```julia
path_A = "/Users/las/Documents/GitHub/REU/Data/data_A.csv"
path_B = "/Users/las/Documents/GitHub/REU/Data/data_B.csv"
df_A = DataFrame(CSV.File(path_A));
df_B = DataFrame(CSV.File(path_B));
```

```julia
count_A = 2*df_A.no;
count_B = 2*df_B.no;
```

```julia
hist_hc = histogram(count_A, bins = 184, label="Hill Climbing Data")
savefig(hist_hc, "hist_hc.png")
```
